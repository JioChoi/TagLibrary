```python
import requests
from bs4 import BeautifulSoup


artists = list()

for i in range(1, 50):
    response = requests.get(f"https://danbooru.donmai.us/tags?commit=Search&page={i}&search%5Bcategory%5D=1&search%5Bhide_empty%5D=yes&search%5Border%5D=count")

    soup = BeautifulSoup(response.content, 'html.parser')
    elements = soup.select("tr .tag-type-1:nth-child(2),.post-count")

    for j in range(0, len(elements), 2):
        artists.append({"artists": elements[j].text, "count": elements[j+1].text})
```

```python
import urllib.request

for i in range(0, len(artists)):
    response = requests.get(f"https://danbooru.donmai.us/posts.json?tags={artists[i]["artists"]}+rating%3Ageneral+order%3Ascore")
    response = response.json()
    print(f"Downloading {artists[i]['artists']}")
    try:
        url = response[0]["media_asset"]["variants"][len(response[0]["media_asset"]["variants"]) - 1]["url"]
        urllib.request.urlretrieve(url, f"artists/jpg/{i}.jpg")
    except Exception as e:
        print(f"ERR on {artists[i]["artists"]}")
        continue

```

```python
import os
from PIL import Image

downloaded = os.listdir("artists/jpg")

listed = []

for i in range(0, len(downloaded)):
    if(downloaded[i] == ".DS_Store"):
        continue
    listed.append(int(downloaded[i].replace(".jpg", "")))

downloaded = sorted(listed)

f = open("temp.txt", "w")

for path in downloaded:
    try:
        img = Image.open(f"artists/jpg/{path}.jpg")
        img.save(f"artists/webp/{path}.webp", "webp", optimize=True, quality=50)
        f.write(f"{{\n\t\"tag\": \"{artists[path]["artists"]}\",\n\t\"filename:\" \"{path}.webp\"\n}},\n")
    except Exception as e:
        print(f"ERR on {path}")
        continue
```
